HERENCIA

    Consiste en tomar todas las características en común y todos los comportamientos en común y englobarlos en una 
    CLASE llamada CLASE PADRE ó SUPERCLASE

Código 1: HERENCIA

```python
# SUPERCLASE
class Vehiculos():
    
    def __init__(self, marca, modelo):
        
        self.marca=marca
        self.modelo=modelo
        self.enmarcha=False
        self.acelera=False
        self.frena=False
        
    def arrancar(self):
        self.enmarcha=True
    
    def acelerar(self):
        self.acelera=True
    
    def frenar(self):
        self.frena=True
    
    def estado(self):
        print ("Marca: ", self.marca, "\nModelo: ", self.modelo, "\nEn Marcha: ",
               self.enmarcha, "\nAcelerando: ", self.acelera, "\nFrenando:" , self.frena)

        
        
        
# Para HEREDAR de una CLASE a otra CLASE, solo hay que poner el nombre de la clase
# heredada en los paréntesis de la nueva CLASE.

class Moto(Vehiculos):
    pass
    
    
    

# En la Instancia debemos pasarle las PROPIEDADES "marca" y "modelo", porque estamos heredando TODO
# incluso el CONSTRUCTOR de la SUPERCLASE "Vehiculos".

miMoto=Moto("Honda", "CBR")

# Con el OBJETO "miMoto" que pertenece en la CLASE "Moto" estamos utilizando: PROPIEDADES y MéTODOS que HEREDAMOS
# de la SUPERCLASE "Vehiculos".

miMoto.estado()
```

```python
# SUPERCLASE
class Vehiculos():
    
    def __init__(self, marca, modelo):
        
        self.marca=marca
        self.modelo=modelo
        self.enmarcha=False
        self.acelera=False
        self.frena=False
        
    def arrancar(self):
        self.enmarcha=True
    
    def acelerar(self):
        self.acelera=True
    
    def frenar(self):
        self.frena=True
    
    def estado(self):
        print ("Marca: ", self.marca, "\nModelo: ", self.modelo, "\nEn Marcha: ",
               self.enmarcha, "\nAcelerando: ", self.acelera, "\nFrenando:" , self.frena)

        
        
        

# Un OBJETO de tipo "Moto" estaría Heredando cinco de Métodos de la SUPERCLASE "Vehiculos": 
# Constructor: "__init__(self, marca, modelo)", arrancar(self), acelerar(self), frenar(self), estado(self)
# y además tiene el Método de la CLASE "Moto": "caballito(self)"
class Moto(Vehiculos):
    hcaballito=""
    
    def caballito(self):
        self.hcaballito="Voy haciendo el caballito."
    
    def estado(self):
        print ("Marca: ", self.marca, "\nModelo: ", self.modelo, "\nEn Marcha: ",
           self.enmarcha, "\nAcelerando: ", self.acelera, "\nFrenando:" , self.frena, "\n", self.hcaballito)
        
miMoto=Moto("Honda", "CBR")

#miMoto.caballito()
miMoto.caballito()

# EL Método "estado" de CLASE "Moto" sobreescribe el Método "estado" de la SUPERCLASE HEREDADA "Vehiculos". 
# Esto se conoce como: SOBREESCRITURA DE MéTODOS.
miMoto.estado()
```

```python
# SUPERCLASE
class Vehiculos():
    
    def __init__(self, marca, modelo):
        
        self.marca=marca
        self.modelo=modelo
        self.enmarcha=False
        self.acelera=False
        self.frena=False
        
    def arrancar(self):
        self.enmarcha=True
    
    def acelerar(self):
        self.acelera=True
    
    def frenar(self):
        self.frena=True
    
    def estado(self):
        print ("Marca: ", self.marca, "\nModelo: ", self.modelo, "\nEn Marcha: ",
               self.enmarcha, "\nAcelerando: ", self.acelera, "\nFrenando:" , self.frena)

class Furgoneta(Vehiculos):
    
    def carga(self, cargar):
        self.cargado=cargar
        if(self.cargado):
            return "La furgoneta está cargada"
        else:
            return "La furgoneta no está cargada"
        
class Moto(Vehiculos):
    hcaballito=""
    
    def caballito(self):
        self.hcaballito="Voy haciendo el caballito"
        
    def estado(self):
        print ("Marca: ", self.marca, "\nModelo: ", self.modelo, "\nEn Marcha: ",
           self.enmarcha, "\nAcelerando: ", self.acelera, "\nFrenando:" , self.frena, "\n", self.hcaballito)
    
class VElectricos(Vehiculos):
    def __init__(self, marca, modelo):
        
        super().__init__(marca, modelo)
        
        self.autonomia=100
    
    def cargarEnergia(self):
        self.cargando=True

        
        
miMoto=Moto("Honda", "CBR")

miMoto.caballito()

miMoto.estado()

miFurgoneta = Furgoneta("Renault", "Kangoo")

miFurgoneta.arrancar()

miFurgoneta.estado()

print(miFurgoneta.carga(True))

class BicicletaElectrica(VElectricos,Vehiculos):
    pass

miBici=BicicletaElectrica("Orbea", "ThJ")
```

```python
# herencia_super.py
# Parte 4: Solución para OBJETO tipo Persona y OBJETO tipo Empleado()



class Persona():

    
    
    def __init__(self, nombre, edad, Lugar_residencia):
        
        self.nombre=nombre
        self.edad=edad
        self.lugar_residencia=Lugar_residencia

        
        
    def descripcion(self):
        
        print("Nombre: ", self.nombre, " Edad: ", self.edad, " Residencia: ", self.lugar_residencia)
        
        
        
class Empleado(Persona):
    
    def __init__(self, salario, antigüedad, nombre_empleado, edad_empleado, residencia_empleado):
        
        
        # La instrucción: super(), esta llamando al Método que se esté indicando enseguida de "super()."  
        # en esté caso la CLASE: "Persona()", porque es de la clase Padre o Superclase y se le están
        #enviando: nombre_empleado, edad_empleado, residencia_empleado.
        
        super().__init__(nombre_empleado, edad_empleado, residencia_empleado)
        self.salario=salario
        self.antigüedad=antigüedad
      
    
    def descripcion(self):
        
        super().descripcion()
        
        print("Salario: ", self.salario, "Antigüedad: ", self.antigüedad)

        
            
Manuel=Empleado(1500, 15, "Manuel", 55, "Colombia")
Manuel.descripcion()
```
